```python
#In probability and statistics, variance is the mean squared deviation of each data point 
#from the mean of the entire dataset.
#Standard deviation (std) is the square root of variance. 

#The mean of [2,3,4] is (2+3+4)/3 = 3.0, and its variation is (2-3)^2 + (3-3)^2 + (4-3)^2 = 1+0+1 = 2. 
#Note that in Python, .var() will return the variance divided by N-1 where N is the length of the data,
#the output is then 2/(3-1) = 1.

# < - .lt()
# <= - .le()
# > - .gt()
# >= - .ge()
# == - .eq()
```
