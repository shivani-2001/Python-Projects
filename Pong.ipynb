```python

# coding: utf-8

# In[ ]:


# Add next level---
import pygame
import random
import numpy as np

pygame.init()
clock = pygame.time.Clock()
speed = 40

display_width = 600
display_height = 600
x = 100
y = 100
radius = 10
dx = 4
dy = 4

paddle_x = 10
paddle_y = 10
paddle_width = 3
paddle_height = 40

play_score = 0

def randomize_start():
    global x, y, dy
    x = random.randint(int(display_width/4), display_width-20)
    y = random.randint(10, display_height-10)
    if random.randint(0, 2)%2 == 0:
        dy *= -1

def hit_back():
    if x+radius > display_width:
        return True
    return False

def hit_sides():
    if y-radius < 0:
        return True
    if y+radius > display_height:
        return True
    return False

def hit_paddle():
    global play_score
    global dx
    global dy
    global paddle_y
    if x-radius <= paddle_x+paddle_width and y > paddle_y and y < paddle_y+paddle_height:
        play_score += 10
        if play_score >= 100 and play_score <= 300:
            dx = 5
            dy = 5
            paddle_y = 7
        if play_score > 300 and play_score <= 600:
            dx = 7
            dy = 7
            paddle_y = 10
        if play_score > 600 and play_score <= 1000:
            dx = 8
            dy = 8
            paddle_y = 12
        return True
    return False

def game_over():
    global play_score
    end_game = True
    
    font_title = pygame.font.Font(None, 30)
    font_instructions = pygame.font.Font(None, 20)
    if play_score >= 1000:
        win = font_title.render("Congratulations!!! You are a great player... Wait for next level", True, (0, 0, 0))
        win_space = win.get_rect(center = (int(display_width/2), int(display_height/3.5)))
        display.blit(win, win_space)
    title = font_title.render("Game over!!!", True, (0, 0, 0))
    title_space = title.get_rect(center = (int(display_width/2), int(display_height/3)))
    display.blit(title, title_space)
    final_score = "Your score : " + str(play_score)
    score = font_instructions.render(final_score, True, (0, 0, 0))
    score_space = score.get_rect(center = (int(display_width/2), int(display_height/2.5)))
    display.blit(score, score_space)
    r_instructions = font_instructions.render("Press R to resume...", True, (0, 0, 0))
    r_space = r_instructions.get_rect(center = (int(display_width/2), int(display_height/2)))
    display.blit(r_instructions, r_space)
    q_instructions = font_instructions.render("Press Q to quit...", True, (0, 0, 0))
    q_space = q_instructions.get_rect(center = (int(display_width/2), int(display_height/1.7)))
    display.blit(q_instructions, q_space)
    
    pygame.display.flip()
    
    while (end_game):
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_q:
                    pygame.quit()
                if event.key == pygame.K_r:
                    return False
    
display = pygame.display.set_mode((display_width, display_height))
pygame.display.set_caption("Hey Buddy!!!")
display.fill((255, 0, 0))
font_welcome = pygame.font.Font(None, 40)
color = np.random.randint(0, 254, 3)
welcome = font_welcome.render("Let's Play!!!", True, (color[0], color[1], color[2]))
welcome_space = welcome.get_rect(center = (int(display_width/2), int(display_height/3)))
display.blit(welcome, welcome_space)
font_start = pygame.font.Font(None, 30)
start = font_start.render("Press Y to start, or autostart in 7 seconds...", True, (0, 0, 255))
start_space = start.get_rect(center = (int(display_width/2), int(display_height/2.3)))
display.blit(start, start_space)
pygame.display.flip()

pygame.time.set_timer(pygame.USEREVENT, 7000)
timer_active = True
while (timer_active):
    for event in pygame.event.get():
        if event.type == pygame.USEREVENT:
            timer_active = False
        if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_y:
                    timer_active = False
        if event.type == pygame.QUIT:
            pygame.quit()
            
randomize_start()
while True:
    clock.tick(speed)
    
    pressed_key = pygame.key.get_pressed()
    if pressed_key[pygame.K_DOWN] or pressed_key[pygame.K_s]:
        if paddle_y+paddle_height+10 <= display_height:
            paddle_y += 6
    if pressed_key[pygame.K_UP] or pressed_key[pygame.K_w]:
        if paddle_y-10 >= 0:
            paddle_y -= 6
            
    display.fill((255, 0, 0))
    x += dx
    y += dy
    
    pygame.draw.rect(display, (0, 0, 255), (paddle_x, paddle_y, paddle_width, paddle_height))
    pygame.draw.circle(display, (0, 255, 0), (x, y), radius)  
    
    if x < radius:
        game_over()
        randomize_start()
        dx = 4
        dy = 4
        paddle_y = 6
        play_score = 0
    if hit_back() or hit_paddle():
        dx *= -1
    if hit_sides():
        dy *= -1
    pygame.display.update()
    for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
```
